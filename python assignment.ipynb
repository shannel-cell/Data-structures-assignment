a) LINEAR SEARCH

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

arr = [21,10,45,32,67,5,18] # Example: Define arr as a list

result = linear_search(arr, 32)
print("Linear Search Result (32 found at index):", result)
```

b) BINARY SEARCH

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    arr = [21, 10, 45, 32, 67, 5, 18]


    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

# Sort list before binary search
sorted_arr = sorted(arr)
print("Sorted array:", sorted_arr)

result = binary_search(sorted_arr, 45)
print("Binary Search Result (45 found at index):", result)
```

c) BUBBLE SORT

```python
def bubble_sort(arr):
    n = len(arr)

    arr = [21, 10, 45, 32, 67, 5, 18]


    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

sorted_list = bubble_sort(arr.copy())
print("Bubble Sorted List:", sorted_list)
```

a)NODE CLASS

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

b) LINKEDLIST CLASS

```python
class LinkedList:
    def __init__(self):
        self.head = None

    # i. Insert at beginning
    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # ii. Insert at end
    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node

    # iii. Display elements
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# Testing LinkedList
ll = LinkedList()
ll.insert_beginning(10)
ll.insert_beginning(5)
ll.insert_end(20)

print("Linked List Elements:")
ll.display()
```

b) PYTHON SIMULATION OF QUEUE

```python
from collections import deque

class StudentQueue:
    def __init__(self):
        self.queue = deque()

    # i. Add student
    def add_student(self, name):
        self.queue.append(name)
        print(f"{name} added to the queue.")

    # ii. Serve student
    def serve_student(self):
        if self.queue:
            served = self.queue.popleft()
            print(f"{served} has been served.")
        else:
            print("No students in queue.")

    # iii. Display queue
    def display_queue(self):
        print("Current Queue:")
        print(list(self.queue))


# Testing
sq = StudentQueue()
sq.add_student("Cate")
sq.add_student("Brian")
sq.add_student("Mark")

sq.display_queue()
sq.serve_student()
sq.display_queue()
```

a) Most Appropriate Data Structure

The most appropriate data structure is a Queue.

Justification:

Students are served in the order they arrive

First student to arrive is served first

This follows the FIFO (First In First Out) principle
